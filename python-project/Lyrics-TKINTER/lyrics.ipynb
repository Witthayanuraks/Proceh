```python
# pip install pillow pygame 

# --------------------- Setup Librarie
import tkinter as tk
import threading
from time import sleep
import itertools
from PIL import Image, ImageTk
import pygame

# -------------------- SETUP MUSIK -------------------- #
pygame.mixer.init()
pygame.mixer.music.load("Robin - Gak Pake Lama.mp3")  # Pastikan file music.mp3 ada
pygame.mixer.music.play(-1)  # Putar musik secara looping

# -------------------- SETUP WINDOW -------------------- #
root = tk.Tk()
root.title("Lirik Animasi dengan Background, Musik & Gambar")

# Load background image
bg_image = Image.open("background.jpg")
bg_photo = ImageTk.PhotoImage(bg_image)

# Tempelkan background pada window
bg_label = tk.Label(root, image=bg_photo)
bg_label.place(x=0, y=0, relwidth=1, relheight=1)

# Atur ukuran window mengikuti background image (atau sesuaikan manual, misalnya "800x600")
root.geometry(f"{bg_image.width}x{bg_image.height}")

# -------------------- LOAD GAMBAR UNTUK LIRIK -------------------- #
# Gambar pertama
img1 = Image.open("./RobinDJ.webp")
img1 = img1.resize((200, 150), Image.ANTIALIAS)  # Resize sesuai kebutuhan
img1_photo = ImageTk.PhotoImage(img1)

# Gambar kedua
img2 = Image.open("./Sticker_PPG_14_Robin_01.webp")
img2 = img2.resize((200, 150), Image.ANTIALIAS)
img2_photo = ImageTk.PhotoImage(img2)

# Simpan referensi agar gambar tidak dihapus oleh garbage collector
root.img1_photo = img1_photo
root.img2_photo = img2_photo

# -------------------- SETUP WIDGET -------------------- #
# Label judul
title_label = tk.Label(root, text="Lirik Animasi Romantis", font=("Helvetica", 16, "bold"),
                       bg="#ffffff", fg="#333333")
title_label.place(x=20, y=20)

# Text widget untuk menampilkan lirik
text_widget = tk.Text(root, font=("Helvetica", 14), width=50, height=15,
                      bg="#ffffff", fg="#000000", bd=5)
text_widget.place(x=20, y=60)

# Dictionary untuk menentukan kapan menampilkan gambar di dalam lirik
# Misal: setelah baris ke-4 (index 3) tampilkan img1, setelah baris ke-8 (index 7) tampilkan img2
images_insertion = {3: img1_photo, 7: img2_photo}

# -------------------- ANIMASI TEKS -------------------- #
def animate_text():
    # List lirik beserta jeda per karakter
    lines = [
        ("Setiap kali kita bertemu", 0.10),
        ("Aku lihat kamu senyum padaku", 0.10),
        ("Ku tahu kamu suka padaku", 0.10),
        ("Tapi kamu ragu bilang begitu", 0.10),
        ("Terlalu lama kamu membuang waktumu", 0.19),
        ("Terlalu lama memendam rasa", 0.22),
        ("Gak pake lama", 0.15),
        ("Bilang saja kalau kau suka", 0.16),
        ("Gak pake lama", 0.15),
        ("Bilang saja kalau kau cinta", 0.16),
        ("Dan aku juga suka", 0.17),
        ("Suka kamu", 0.14),
        ("Ku tunggu sampai kamu bilang I love you", 0.17),
        ("❤❤❤❤❤❤❤❤❤❤❤❤❤❤❤❤❤❤❤❤❤❤❤❤", 0.15)
    ]
    
    # Siapkan warna untuk tiap baris secara bergiliran
    colors = ["red", "green", "blue", "purple", "orange", "cyan"]
    color_cycle = itertools.cycle(colors)
    
    for index, (line, delay) in enumerate(lines):
        current_color = next(color_cycle)
        # Konfigurasi tag untuk mewarnai teks
        text_widget.tag_config(current_color, foreground=current_color)
        
        # Tampilkan tiap karakter dengan delay
        for char in line:
            text_widget.insert(tk.END, char, current_color)
            text_widget.update()
            sleep(delay)
        text_widget.insert(tk.END, "\n")
        
        # Jika indeks baris sesuai, masukkan gambar ke dalam Text widget
        if index in images_insertion:
            text_widget.image_create(tk.END, image=images_insertion[index])
            text_widget.insert(tk.END, "\n")
    
    # Setelah animasi selesai, aktifkan kembali tombol Mulai
    start_button.config(state=tk.NORMAL)

def start_animation():
    start_button.config(state=tk.DISABLED)
    threading.Thread(target=animate_text, daemon=True).start()

def close_app():
    pygame.mixer.music.stop()
    root.destroy()

# -------------------- TOMBOL -------------------- #
start_button = tk.Button(root, text="Mulai", command=start_animation)
start_button.place(x=20, y=420)
exit_button = tk.Button(root, text="Keluar", command=close_app)
exit_button.place(x=80, y=420)

root.mainloop()
```
